```python
# Import dependencies

import pandas as pd
import psychrolib
import os
import json
```

```python
os.chdir('C:/Users/amitc_crl/OneDrive/Documents/GitHub/ClimateData/Dehumidification')
psychrolib.SetUnitSystem(psychrolib.IP)
weatherFile = 'Illinois, Chicago-Midway AP (725340) - PHIUS+ Climate Data.xls'

data = pd.read_excel(str(weatherFile))
location = data['Passive House Planning'].values[4]
altitude = data['Unnamed: 7'].values[19]
altitude = 0
atmPressure = psychrolib.GetStandardAtmPressure(altitude)
humRatioLimit = 0.01

days = []
dewPoints = []
humRatios = []
deltas = []
ddds = []

for x in range(12):
    n = x + 2
    column = 'Unnamed: ' + str(n)
    day = data[str(column)].values[3]
    dewPoint = data[str(column)].values[26]
    days.append(day)
    dewPoints.append(dewPoint)

for x in dewPoints:
    humRatio = psychrolib.GetHumRatioFromTDewPoint(x,atmPressure)
    humRatios.append(humRatio)
    delta = humRatio - humRatioLimit
    if delta > 0:
        deltas.append(delta)
    else:
        deltas.append(0)

for x in range(len(days)):
    ddd = days[x] * deltas[x]
    ddds.append(ddd)

ddd = sum(ddds)
print(ddd)

```
